```python
# 00_configure_storage
storage_account = "ecomdatalakegen2"      # your real storage account name
storage_key = dbutils.secrets.get(scope="azure", key="storage-key")

spark.conf.set(
    f"fs.azure.account.key.{storage_account}.dfs.core.windows.net",
    storage_key
)

print("Storage key configured successfully.")
```
